```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(11)
```

```python
x
```

```python
y=x*x
```

```python
y
```

```python
print(x,y)
```

### Scatter Plot

```python
plt.scatter(x,y)
```

### Add Styling // Plot

```python
plt.plot(x,y,'r-*')  #line color r, Line Style - , Line Symbol *
```

### Add Label

```python
plt.plot(x,y,'r-*')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Figure Title")
plt.show()
```

### Change Range of Axis // Xticks

```python
plt.plot(x,y,'r-*')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Figure Title")
plt.xticks(range(0,11))
plt.yticks(range(0,101,10))

plt.show()
```

### Multiple Graph //Subplot

```python
plt.subplot(3,3,1)
plt.plot(x,y, 'r-*')

plt.subplot(2,3,2)
plt.plot(x,y, 'g-+')

plt.subplot(1,3,3)
plt.plot(x,y, 'b--')
```

#### Save as

```python
fig=plt.figure(figsize=(10,5))
#(10,5) is height width ratio

axes=fig.add_axes([0.1,0.2,0.8,0.3])
# Space from each side
#left, bottom, width, height 0.1,0.2,0.8,0.3

axes. plot(x,y,'r-*')
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Test Figure")

fig.savefig("graph.pdf", dpi=200)
```

### Mutiple lines on same figure

```python
fig,axes=plt.subplots(figsize=(10,8))
axes.plot(x,x+1,color='Red',linewidth=0.25)
axes.plot(x,x+2,color='Red',linewidth=0.50)
axes.plot(x,x+3,color='Red',linewidth=1.00)
axes.plot(x,x+4,color='Red',linewidth=1.25)

axes.plot(x,x+5,color='Green',lw=1.25, linestyle='-')
axes.plot(x,x+6,color='Green',lw=1.25, linestyle='-.')
axes.plot(x,x+7,color='Green',lw=1.25, linestyle=':')


axes.plot(x,x+8,color='Blue',lw=3, ls='--',marker='o',markersize='8')


axes.plot(x,x+9,color='Blue',lw=3, ls='--',marker='s',markersize=8,markerfacecolor='red',markeredgecolor='green',markeredgewidth='3')

axes.plot(x,x+10,color='Blue',lw=3, ls='--',marker='1',markersize=8,markeredgecolor='green',markeredgewidth='3')
```
