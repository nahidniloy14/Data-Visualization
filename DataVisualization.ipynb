```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(11)
```

```python
x
```

```python
y=x*x
```

```python
y
```

```python
print(x,y)
```

### Scatter Plot

```python
plt.scatter(x,y)
```

### Add Styling // Plot

```python
plt.plot(x,y,'r-*')  #line color r, Line Style - , Line Symbol *
```

### Add Label

```python
plt.plot(x,y,'r-*')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Figure Title")
plt.show()
```

### Change Range of Axis // Xticks

```python
plt.plot(x,y,'r-*')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Figure Title")
plt.xticks(range(0,11))
plt.yticks(range(0,101,10))

plt.show()
```
